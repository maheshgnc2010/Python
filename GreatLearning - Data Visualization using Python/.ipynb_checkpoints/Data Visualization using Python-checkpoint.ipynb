```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline # help to remove plt.show()
sns.set(color_codes=True)
```

```python
auto =pd.read_csv('Automobile.csv')
```

```python
auto.head()
```

```python
auto.shape
```

#### Univariate Analysis

- It is referred as analysis single variables statistics or distrubution etc and getting meaning out of it.

```python
# Distance/Distribution Plot or Histogram  -  method works on only with numerical data
sns.distplot(auto['normalized_losses']) # it includes kernal density line 
plt.show()
```

```python
sns.distplot(auto['normalized_losses'], kde=False, rug=True) # here we are removing kde line and include freq count with rug 
plt.show()
```

```python
# Distance Plot or Histogram 
sns.displot(auto['normalized_losses'])
plt.show() # we need to use plt as seaborn is based out of matplotlib
```

#### Bivariate Analysis

- Trying to understand pattern or underlying strucuture between two variables

1. jointplot 
- we use this method to build different kind of graphs to show relationship between two variables. - kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
- It is the combination of two graphs the outer graph with histogram or histogram with kde 

```python
# jointplot helps to understand correlation between the two variables 
sns.jointplot(auto['engine_size'],auto['horsepower'])
plt.show()
```

```python
# "hex" plot
sns.jointplot(auto['engine_size'],auto['horsepower'], kind="hex")
plt.show()
```

```python
# kde plot 
sns.jointplot(auto['engine_size'],auto['horsepower'], kind='kde')
```

```python
# hist plot 
sns.jointplot(auto['engine_size'],auto['horsepower'], kind='hist')
```

```python
# regression plot 
sns.jointplot(auto['engine_size'],auto['horsepower'], kind='reg')
```

```python
# residual plot 
sns.jointplot(auto['engine_size'],auto['horsepower'], kind='resid')
```

#### Multivariate Analysis

- Helps to understand underlying relationship or pattern between the multiple variable

1. pairplot - Plot pairwise relationships in a dataset.

- kind : {'scatter', 'kde', 'hist', 'reg'}

```python
sns.pairplot(auto[['normalized_losses','engine_size','horsepower']])
```
